```python
import pandas as pd
```

```python
df = pd.read_csv("../source/Ethnicity.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[df1['Cohort'] == '4 year June']
df2
```

```python
school_names = df2[['DBN', 'School Name', 'Cohort Year']].drop_duplicates()
school_names
```

```python
asian = df2.loc[df2['Category'] == 'Asian', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
asian = asian.add_prefix('Asian ').reset_index()
asian = asian.rename({'Asian Cohort Year': 'Cohort Year'}, axis=1)
asian
```

```python
black = df2.loc[df2['Category'] == 'Black', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
black = black.add_prefix('Black ').reset_index()
black = black.rename({'Black Cohort Year': 'Cohort Year'}, axis=1)
black
```

```python
hispanic = df2.loc[df2['Category'] == 'Hispanic', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
hispanic = hispanic.add_prefix('Hispanic ').reset_index()
hispanic = hispanic.rename({'Hispanic Cohort Year': 'Cohort Year'}, axis=1)
hispanic
```

```python
white = df2.loc[df2['Category'] == 'White', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
white = white.add_prefix('White ').reset_index()
white = white.rename({'White Cohort Year': 'Cohort Year'}, axis=1)
white
```

```python
native_american = df2.loc[df2['Category'] == 'Native American', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
native_american = native_american.add_prefix('Native American ').reset_index()
native_american = native_american.rename({'Native American Cohort Year': 'Cohort Year'}, axis=1)
native_american
```

```python
multi_racial = df2.loc[df2['Category'] == 'Multi-Racial', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
multi_racial = multi_racial.add_prefix('Multi-Racial ').reset_index()
multi_racial = multi_racial.rename({'Multi-Racial Cohort Year': 'Cohort Year'}, axis=1)
multi_racial
```

```python
ethnicities = school_names.merge(asian, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities = ethnicities.merge(black, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities = ethnicities.merge(hispanic, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities = ethnicities.merge(white, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities = ethnicities.merge(native_american, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities = ethnicities.merge(multi_racial, how = 'left', on = ['DBN', 'Cohort Year'])
ethnicities
```

```python
ethnicities.to_csv("transformed_ethnicities_table.csv", index=False)
```
