```python
import pandas as pd
```

```python
# Import data sets
class_df = pd.read_csv("school_class.csv")
all_df = pd.read_csv("transformed_all_table.csv")
```

```python
class_df
```

```python
all_df
```

```python
all_short_df = all_df.drop(['% Grads', '% Total Regents of Cohort', '% Advanced Regents of Cohort', '% Dropout'], 1)
all_short_df
```

```python
eth_df = pd.read_csv("transformed_ethnicities_table.csv")
eth_df
```

```python
eth_short_df = eth_df[['DBN', 'School Name', 'Cohort Year', 'Asian # Total Cohort', 'Black # Total Cohort', 'Hispanic # Total Cohort', 'White # Total Cohort', 'Native American # Total Cohort', 'Multi-Racial # Total Cohort']]
eth_short_df
```

```python
gender_df = pd.read_csv("transformed_gender_table.csv")
```

```python
gender_short_df = gender_df[['DBN', 'School Name', 'Cohort Year', 'Female # Total Cohort', 'Male # Total Cohort']]
gender_short_df
```

```python
poverty_df = pd.read_csv("transformed_poverty_table.csv")
poverty_df
```

```python
poverty_short_df = poverty_df[['DBN', 'School Name', 'Cohort Year', 'Econ Disadv # Total Cohort', 'Not Econ Disadv # Total Cohort']]
poverty_short_df
```

```python
swd_df = pd.read_csv("transformed_swd_table.csv")
swd_df
```

```python
swd_short_df = swd_df[['DBN', 'School Name', 'Cohort Year', 'SWD # Total Cohort', 'Not SWD # Total Cohort']]
swd_short_df
```

```python
df_merged = pd.concat([all_short_df, eth_short_df, gender_short_df, poverty_short_df, swd_short_df], join='inner', axis=1)
df_merged
```

```python
df_merged = df_merged.loc[:,~df_merged.columns.duplicated()]
```

```python
s_list = df_merged[df_merged['# Grads']=='s'].index.tolist()
```

```python
clean_df_merged = df_merged.drop(s_list, axis=0)
clean_df_merged
```

```python
clean_df_merged = clean_df_merged.fillna(0)
```

```python
clean_df_merged = clean_df_merged.rename(columns={"Asian # Total Cohort": "# Asian", "Black # Total Cohort": "# Black", "Hispanic # Total Cohort": "# Hispanic", "White # Total Cohort": "# White", "Native American # Total Cohort": "# Native American", "Multi-Racial # Total Cohort": "# Multi-Racial", "Female # Total Cohort": "# Female", "Male # Total Cohort": "# Male", "Econ Disadv # Total Cohort": "# Econ Disadv", "Not Econ Disadv # Total Cohort": "# Not Econ Disadv", "SWD # Total Cohort": "# SWD", "Not SWD # Total Cohort": "# Not SWD"})
clean_df_merged
```

```python
clean_df_merged = clean_df_merged.set_index("School Name")
clean_df_merged
```

```python
clean_df_merged.dtypes
```

```python
columns = list(clean_df_merged.columns)       
for col in columns: 
    if "#" in col: 
        clean_df_merged[col] = clean_df_merged[col].astype('int64')
```

```python
clean_df_merged.dtypes
```

```python
clean_df_merged.to_csv("total_school_data_cohort.csv", index=False)
```

```python
Total_cohorts = clean_df_merged.groupby('School Name')['# Total Cohort'].sum()
Grads = clean_df_merged.groupby('School Name')['# Grads'].sum()
Total_Regents = clean_df_merged.groupby('School Name')['# Total Regents'].sum()
Advanced_Regents = clean_df_merged.groupby('School Name')['# Advanced Regents'].sum()
Dropout = clean_df_merged.groupby('School Name')['# Dropout'].sum()
Asian = clean_df_merged.groupby('School Name')['# Asian'].sum()
Black = clean_df_merged.groupby('School Name')['# Black'].sum()
Hispanic = clean_df_merged.groupby('School Name')['# Hispanic'].sum()
White = clean_df_merged.groupby('School Name')['# White'].sum()
Native_American = clean_df_merged.groupby('School Name')['# Native American'].sum()
Multi_Racial = clean_df_merged.groupby('School Name')['# Multi-Racial'].sum()
Female = clean_df_merged.groupby('School Name')['# Female'].sum()
Male = clean_df_merged.groupby('School Name')['# Male'].sum()
Econ_Disadv = clean_df_merged.groupby('School Name')['# Econ Disadv'].sum()
Not_Econ_Disadv = clean_df_merged.groupby('School Name')['# Not Econ Disadv'].sum()
SWD = clean_df_merged.groupby('School Name')['# SWD'].sum()
Not_SWD = clean_df_merged.groupby('School Name')['# Not SWD'].sum()
Total_cohorts
```

```python
total_school_data = pd.concat([Total_cohorts, Grads, Total_Regents, Advanced_Regents, Dropout, Female, Male, White, Black, Asian, Hispanic, Native_American, Multi_Racial, Econ_Disadv, Not_Econ_Disadv, SWD, Not_SWD], join='inner', axis=1)
total_school_data
```
