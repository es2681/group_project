```python
import pandas as pd
```

```python
df = pd.read_csv("../source/Poverty.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[df1['Cohort'] == '4 year June']
df2
```

```python
school_names = df2[['DBN', 'School Name', 'Cohort Year']].drop_duplicates()
school_names
```

```python
econ_disadv = df2.loc[df2['Category'] == 'Econ Disadv', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
econ_disadv  = econ_disadv.add_prefix('Econ Disadv ').reset_index()
econ_disadv = econ_disadv.rename({'Econ Disadv Cohort Year': 'Cohort Year'}, axis=1)
econ_disadv 
```

```python
not_econ_disadv = df2.loc[df2['Category'] == 'Not Econ Disadv', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
not_econ_disadv  = not_econ_disadv.add_prefix('Not Econ Disadv ').reset_index()
not_econ_disadv = not_econ_disadv.rename({'Not Econ Disadv Cohort Year': 'Cohort Year'}, axis=1)
not_econ_disadv 
```

```python
poverty = school_names.merge(econ_disadv , how = 'left', on = ['DBN', 'Cohort Year'])
poverty = poverty.merge(not_econ_disadv, how='left', on=['DBN', 'Cohort Year'])
poverty
```

```python
poverty.to_csv("transformed_poverty_table.csv", index=False)
```
