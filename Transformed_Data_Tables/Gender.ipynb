```python
import pandas as pd
```

```python
df = pd.read_csv("../source/Gender.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[df1['Cohort'] == '4 year June']
df2
```

```python
school_names = df2[['DBN', 'School Name', 'Cohort Year']].drop_duplicates()
school_names
```

```python
female = df2.loc[df2['Category'] == 'Female', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
female = female.add_prefix('Female ')
female = female.rename({'Female Cohort Year': 'Cohort Year'}, axis=1)
female
```

```python
male = df2.loc[df2['Category'] == 'Male', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
male = male.add_prefix('Male ')
male = male.rename({'Male Cohort Year': 'Cohort Year'}, axis=1)
male
```

```python
gender = school_names.merge(female, how = 'left', on = ['DBN', 'Cohort Year'])
gender = gender.merge(male, how = 'left', on = ['DBN', 'Cohort Year'])
gender
```

```python
gender.to_csv("transformed_gender_table.csv", index=False)
```
