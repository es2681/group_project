```python
import pandas as pd
```

```python
df = pd.read_csv("../source/Ethnicity.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[(df1['Cohort'] == '4 year June') & (df1['Cohort Year'] == 2017)]
df2
```

```python
df2['Category'].value_counts()
school_names = df2[['DBN', 'School Name']].drop_duplicates()
school_names
```

```python
asian = df2.loc[df2['Category'] == 'Asian', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
asian = asian.add_prefix('Asian ').reset_index()
asian
```

```python
black = df2.loc[df2['Category'] == 'Black', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
black = black.add_prefix('Black ').reset_index()
black
```

```python
hispanic = df2.loc[df2['Category'] == 'Hispanic', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
hispanic = hispanic.add_prefix('Hispanic ').reset_index()
hispanic
```

```python
white = df2.loc[df2['Category'] == 'White', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
white = white.add_prefix('White ').reset_index()
white
```

```python
native_american = df2.loc[df2['Category'] == 'Native American', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
native_american = native_american.add_prefix('Native American ').reset_index()
native_american
```

```python
multi_racial = df2.loc[df2['Category'] == 'Multi-Racial', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
multi_racial = multi_racial.add_prefix('Multi-Racial ').reset_index()
multi_racial
```

```python
ethnicities = school_names.merge(asian, how = 'left', on = 'DBN')
ethnicities = ethnicities.merge(black, how = 'left', on = 'DBN')
ethnicities = ethnicities.merge(hispanic, how = 'left', on = 'DBN')
ethnicities = ethnicities.merge(white, how = 'left', on = 'DBN')
ethnicities = ethnicities.merge(native_american, how = 'left', on = 'DBN')
ethnicities = ethnicities.merge(multi_racial, how = 'left', on = 'DBN')
ethnicities
```
