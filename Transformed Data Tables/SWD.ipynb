```python
import pandas as pd
```

```python
df = pd.read_csv("../source/SWD.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[(df1['Cohort'] == '4 year June') & (df1['Cohort Year'] == 2017)]
df2
```

```python
df2['Category'].value_counts()
school_names = df2[['DBN', 'School Name']].drop_duplicates()
school_names
```

```python
swd_df = df2.loc[df2['Category'] == 'SWD', ['DBN', '# Grads', '% Grads', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
swd_df  = swd_df.add_prefix('SWD ').reset_index()
swd_df
```

```python
swd = school_names.merge(swd_df, how = 'left', on = 'DBN')
swd
```
