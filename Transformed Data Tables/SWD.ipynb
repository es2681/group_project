```python
import pandas as pd
```

```python
df = pd.read_csv("../source/SWD.csv")
df
```

```python
column_list = ['DBN', 'School Name', 'Cohort Year', 'Cohort', 'Category', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']
column_list
```

```python
df1 = df[column_list].copy()
df1
```

```python
df2 = df1.loc[df1['Cohort'] == '4 year June']
df2
```

```python
school_names = df2[['DBN', 'School Name', 'Cohort Year']].drop_duplicates()
school_names
```

```python
swd_df = df2.loc[df2['Category'] == 'SWD', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
swd_df  = swd_df.add_prefix('SWD ').reset_index()
swd_df = swd_df.rename({'SWD Cohort Year': 'Cohort Year'}, axis=1)
swd_df
```

```python
not_swd_df = df2.loc[df2['Category'] == 'Not SWD', ['DBN', 'Cohort Year', '# Total Cohort', '# Grads', '% Grads', '# Total Regents', '% Total Regents of Cohort', '# Advanced Regents', '% Advanced Regents of Cohort', '# Dropout', '% Dropout']].set_index('DBN')
not_swd_df  = not_swd_df.add_prefix('Not SWD ').reset_index()
not_swd_df = not_swd_df.rename({'Not SWD Cohort Year': 'Cohort Year'}, axis=1)
not_swd_df
```

```python
swd = school_names.merge(swd_df, how = 'left', on = ['DBN', 'Cohort Year'])
swd = swd.merge(not_swd_df, how='left', on = ['DBN', 'Cohort Year'])
swd
```

```python
swd.to_csv("transformed_swd_table.csv", index=False)
```
