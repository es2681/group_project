```python
import pandas as pd
```

```python
all_df = pd.read_csv("transformed_all_table.csv")
all_df
```

```python
all_short_df = all_df.drop(['# Grads', '# Total Regents', '# Advanced Regents', '# Dropout'], 1)
all_short_df
```

```python
eth_df = pd.read_csv("transformed_ethnicities_table.csv")
eth_df
```

```python
eth_df.columns
```

```python
eth_short_df = eth_df[['DBN', 'School Name', 'Cohort Year', 'Asian # Total Cohort', 'Black # Total Cohort', 'Hispanic # Total Cohort', 'White # Total Cohort', 'Native American # Total Cohort', 'Multi-Racial # Total Cohort']]
eth_short_df
```

```python
gender_df = pd.read_csv("transformed_gender_table.csv")
gender_df
```

```python
gender_df.columns
```

```python
gender_short_df = gender_df[['DBN', 'School Name', 'Cohort Year', 'Female # Total Cohort', 'Male # Total Cohort']]
gender_short_df
```

```python
poverty_df = pd.read_csv("transformed_poverty_table.csv")
poverty_df
```

```python
poverty_df.columns
```

```python
poverty_short_df = poverty_df[['DBN', 'School Name', 'Cohort Year', 'Econ Disadv # Total Cohort', 'Not Econ Disadv # Total Cohort']]
poverty_short_df
```

```python
swd_df = pd.read_csv("transformed_swd_table.csv")
swd_df
```

```python
swd_df.columns
```

```python
swd_short_df = swd_df[['DBN', 'School Name', 'Cohort Year', 'SWD # Total Cohort', 'Not SWD # Total Cohort']]
swd_short_df
```

```python
df_merged = pd.concat([all_short_df, eth_short_df, gender_short_df, poverty_short_df, swd_short_df], join='inner', axis=1)
df_merged
```

```python
df_merged.columns
```

```python
df_merged = df_merged.loc[:,~df_merged.columns.duplicated()]
```

```python
df_merged.columns
```

```python
s_list = df_merged[df_merged['% Grads']=='s'].index.tolist()
```

```python
clean_df_merged = df_merged.drop(s_list, axis=0)
clean_df_merged
```

```python
clean_df_merged.shape
```

```python
clean_df_merged.isnull().sum()
```

```python
clean_df_merged = clean_df_merged.fillna(0)
```

```python
clean_df_merged.isnull().sum()
```

```python
clean_df_merged.dtypes
```

```python
columns = list(clean_df_merged.columns)
for col in columns:
    if "%" in col:
        clean_df_merged[col] = clean_df_merged[col].astype('float')
        
for col in columns: 
    if "#" in col: 
        clean_df_merged[col] = clean_df_merged[col].astype('int64')
```

```python
clean_df_merged.dtypes
```

```python
clean_df_merged['% Asian Cohort'] = (clean_df_merged['Asian # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Black Cohort'] = (clean_df_merged['Black # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Hispanic Cohort'] = (clean_df_merged['Hispanic # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% White Cohort'] = (clean_df_merged['White # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Native American Cohort'] = (clean_df_merged['Native American # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Multi-Racial Cohort'] = (clean_df_merged['Multi-Racial # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Female Cohort'] = (clean_df_merged['Female # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Male Cohort'] = (clean_df_merged['Male # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Econ Disadv Cohort'] = (clean_df_merged['Econ Disadv # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Not Econ Disadv Cohort'] = (clean_df_merged['Not Econ Disadv # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% SWD Cohort'] = (clean_df_merged['SWD # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged['% Not SWD Cohort'] = (clean_df_merged['Not SWD # Total Cohort'] / clean_df_merged['# Total Cohort']) * 100
clean_df_merged
```

```python
clean_df_merged = clean_df_merged.drop(['# Total Cohort', 'Asian # Total Cohort', 'Black # Total Cohort',
       'Hispanic # Total Cohort', 'White # Total Cohort',
       'Native American # Total Cohort', 'Multi-Racial # Total Cohort',
       'Female # Total Cohort', 'Male # Total Cohort',
       'Econ Disadv # Total Cohort', 'Not Econ Disadv # Total Cohort',
       'SWD # Total Cohort', 'Not SWD # Total Cohort',], 1)
```

```python
columns = list(clean_df_merged.columns)
for col in columns: 
    if clean_df_merged[col].dtypes == 'float64':
        clean_df_merged.round(1)
        
clean_df_merged
```

```python
clean_df_merged = clean_df_merged.round(1)
clean_df_merged
```

```python
clean_df_merged['School Name and Cohort Year'] = clean_df_merged['School Name'] + clean_df_merged['Cohort Year'].astype(str)
clean_df_merged.head()
```

```python
clean_df_merged.columns = clean_df_merged.columns.str.replace('%', 'Percent')
clean_df_merged.columns = clean_df_merged.columns.map(lambda x : x.replace("_", "-").replace("-", ""))
clean_df_merged.columns
```

```python
clean_df_merged.to_csv("all_data_concat.csv", index=False)
```
