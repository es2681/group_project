```python
import pandas as pd
import numpy as np
from pathlib import Path  
```

```python
# Load the dataset disadvantage
file_path = "/Users/pace/Desktop/Columbia Bootcamp/Final Project/transformed_poverty_disadv_table.csv"
transformed_poverty_disadv_df = pd.read_csv(file_path)
transformed_poverty_disadv_df
```

```python
school_names = transformed_poverty_disadv_df[['dbn', 'school_name', 'cohort_year']].drop_duplicates()
school_names
```

```python
# Load the dataset not disadvantage
file_path = "/Users/pace/Desktop/Columbia Bootcamp/Final Project/transformed_poverty_not_disadv_table.csv"
transformed_poverty_not_disadv_df = pd.read_csv(file_path)
transformed_poverty_not_disadv_df
```

```python
poverty = school_names.merge(transformed_poverty_disadv_df , how = 'right', on = ['dbn', 'cohort_year'])
poverty = poverty.merge(transformed_poverty_not_disadv_df, how='left', on=['dbn', 'cohort_year'])
poverty
```

```python
poverty.drop(['school_name_y','school_name'], axis =1)
```

```python
poverty.to_csv("/Users/pace/Desktop/Columbia Bootcamp/Final Project/transformed_poverty_all_table.csv", index=False)
```
