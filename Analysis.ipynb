# Data Analysis

```python
import pandas as pd
```

```python
# Import main dataset
df = pd.read_csv("Transformed_Data_Tables/all_school_data_with_categories.csv")
```

```python
# Filter by Year_Cohort >= 2014
df = df[df['Cohort Year'] >= 2014]
df.head()
```

```python
# Drop DBN Column
df = df.loc[:, df.columns != 'DBN']
df.head()
```

```python
# Sort by School Name and Year
df.sort_values(by=['School Name', 'Cohort Year'], inplace=True)
df.head()
```

```python
# Reset Index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# Save in JSON Format
file_name = f'json/schools_2014-2017.json'
df.to_json(file_name)
```

```python
# Import complementary dataset (Public Schools Data)
df_public = pd.read_csv("source/Public_Schools.csv")
df_public.head()
```

```python
df_public = df_public[['NAME', 'CITY', 'ZIP', 'POPULATION', 'COUNTY', 'LATITUDE', 'LONGITUDE']]
df_public.head()
```

```python
# Rename Name Column
df_public = df_public.rename(columns={'NAME': 'School Name'})
df_public.head()
```

```python
# 320 schools (100% match)
df_merged = df.merge(df_public, left_on='School Name', right_on='School Name')
df_merged.head()
```
